```python
import numpy as np
import pandas as pd
import nibabel as nib
import glob
import os
from radiomics import featureextractor
from math import isclose
from scipy import ndimage, spatial
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from nibabel.affines import apply_affine
%matplotlib widget
import matplotlib.pyplot as plt


def fromspacetovox(inv_affine, where_space):

    mask_vox = (np.dot(inv_affine[:3, :3], where_space.T) + inv_affine[:3, 3:4]).T
    return mask_vox


def fromvoxtospace(affine, where_ones):

    mask_space = apply_affine(affine, where_ones)
    return mask_space


def load_nib_file(path):
    mask = nib.load(path)
    mask_data = mask.get_fdata()

    return mask, mask_data


def compute_dice_score(agg, gt):
    dice = 1 - spatial.distance.dice(agg.flatten(), gt.flatten())
    return dice


path_data = "/mnt/data/psteinmetz/neotex/data_MRI/2training_sets_forPCA/"
path_data_test = "/mnt/data/psteinmetz/neotex/data_MRI/follow_up/Testing/"
path_data_sheet = "/mnt/data/psteinmetz/neotex/data_MRI/results_Training+Ozgun_test_BDD_25.ods"

score = pd.read_excel(path_data_sheet, sheet_name="Sheet2", index_col='case')

list_training = [x for x in os.listdir(path_data) if '.' not in x]
train_features = {}
train_shape_features = {}
list_testing = [x for x in os.listdir(path_data_test) if '.' not in x]
test_features = {}
test_shape_features = {}
errors = []
images_training = glob.glob(path_data + '/*/GT_RS_CR_2mm.nii.gz')
images_testing = glob.glob(path_data_test + '/*/GT_RS_CR_2mm.nii.gz')
centered_segmentations_train = {}
centered_segmentations_test = {}

for case in list_training + list_testing:
    print(case)

    if case in list_training:
        path = path_data
    elif case in list_testing:
        path = path_data_test

    image = glob.glob(path + case + '/T1_*_CR_2mm.nii.gz')
    mask = glob.glob(path + case + '/GT_RS_CR_2mm.nii.gz')

    try:
        features = featureextractor.RadiomicsFeatureExtractor()
        features.disableAllFeatures()
        features.enableFeatureClassByName('shape')
        results = features.execute(image[0], mask[0])
        features = {key: float(results[key]) for key in results.keys() if 'shape' in key}
        if case in list_training:
            train_features.update({case: results})
            train_shape_features.update({case: features})
        elif case in list_testing:
            test_features.update({case: results})
            test_shape_features.update({case: features})
    except ValueError:
        errors.append(mask)


# fit scaler + PCA model
# scaler3 = StandardScaler().fit(pd.concat((pd.DataFrame.from_dict(train_shape_features).T,
#                                          pd.DataFrame.from_dict(test_shape_features).T)))
scaler3 = StandardScaler().fit(pd.DataFrame.from_dict(train_shape_features).T)
X_train_scaled3 = scaler3.transform(pd.DataFrame.from_dict(train_shape_features).T)
X_test_scaled3 = scaler3.transform(pd.DataFrame.from_dict(test_shape_features).T)

pca3 = PCA(n_components=3)
pca3.fit(np.concatenate((X_train_scaled3, X_test_scaled3)))
explained_variance_ratio3 = sum(pca3.explained_variance_ratio_)
X_train_PCA_projected3 = pca3.transform(X_train_scaled3)


# test transform
X_test_PCA_projected3 = pca3.transform(X_test_scaled3)

dist_same_scale_same_pca = spatial.distance.cdist(X_test_scaled3, X_train_scaled3, 'euclidean').T
distances_same_scale_same_pca = pd.DataFrame(dist_same_scale_same_pca, index=train_features.keys(), columns=test_features.keys())

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X_train_PCA_projected3[:, 0], X_train_PCA_projected3[:, 1], X_train_PCA_projected3[:, 2], c='b', marker='o')
ax.scatter(X_test_PCA_projected3[:, 0], X_test_PCA_projected3[:, 1], X_test_PCA_projected3[:, 2], c='r', marker='^')
ax.set_xlim3d(min(min(X_train_PCA_projected3[:, 0]), min(X_test_PCA_projected3[:, 0])),
              max(max(X_train_PCA_projected3[:, 0]), max(X_test_PCA_projected3[:, 0])))
ax.set_ylim3d(min(min(X_train_PCA_projected3[:, 1]), min(X_test_PCA_projected3[:, 1])),
              max(max(X_train_PCA_projected3[:, 1]), max(X_test_PCA_projected3[:, 1])))
ax.set_zlim3d(min(min(X_train_PCA_projected3[:, 2]), min(X_test_PCA_projected3[:, 2])),
              max(max(X_train_PCA_projected3[:, 2]), max(X_test_PCA_projected3[:, 2])))
ax.set_xlabel('1st component')
ax.set_ylabel('2nd component')
ax.set_zlabel('3rd component')
plt.show()

knn_10 = distances_same_scale_same_pca.transform(np.sort).reset_index(drop=True)
stats = knn_10.mean()
results_to_plot = pd.concat((stats, score['2trainsets']), axis=1).dropna()
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(results_to_plot[0], results_to_plot['2trainsets'], c='b', marker='o')
ax.set_xlim(3.3, 6)
ax.set_xlabel('Average 3D distance between test data instance and training set')
ax.set_ylabel('Dice score vs GT')
ax.set_title('Prediction results vs Data instance similarity w/ train set')
plt.show()

summary3 = distances_same_scale_same_pca.describe()
third_test = summary3.idxmax(axis="columns")

# for mask in images_training + images_testing:
#
#     computation = {}
#     gt, gt_data = load_nib_file(mask)
#     ones = np.argwhere(gt_data == 1)
#
#     bary = ndimage.center_of_mass(gt_data)
#     center_image = (75, 40, 50)
#
#     aff = gt.affine
#     center_mask_space = fromvoxtospace(aff, bary)
#     center_image_space = fromvoxtospace(aff, center_image)
#     gt_data_space = fromvoxtospace(aff, ones)
#
#     translation = -(center_mask_space-center_image_space)
#     translated_gt_data_space = gt_data_space + translation
#
#     inv_aff = np.linalg.inv(aff)
#     gt_vox = np.rint(fromspacetovox(inv_aff, translated_gt_data_space)).astype(int)
#     new_mask = np.zeros((150, 80, 100))
#     new_mask[gt_vox[:, 0], gt_vox[:, 1], gt_vox[:, 2]] = 1
#     new_bary = ndimage.center_of_mass(new_mask)
#
#     computation.update({'center_image_coordinates': new_bary})
#     computation.update({'centered_mask_GT': new_mask})
#     computation.update({'affine': aff})
#     computation.update({'inv_affine': inv_aff})
#
#     if mask in images_training:
#         centered_segmentations_train.update({mask.split('/')[-2]: computation})
#     elif mask in images_testing:
#         centered_segmentations_test.update({mask.split('/')[-2]: computation})
#
# dices_glob = {}
# comb = []
# for key, mask in centered_segmentations_test.items():
#     dices = {}
#     print(key)
#     for sub_key, sub_mask in centered_segmentations_train.items():
#
#         dice = compute_dice_score(mask['centered_mask_GT'], sub_mask['centered_mask_GT'])
#         print('dice: ' + str(dice))
#         dices.update({sub_key: dice})
#
#     dices_glob.update({key: dices})
# results_glob = pd.DataFrame.from_dict(dices_glob)
# results_glob.to_csv(path_data + r'/results_25MP.csv')
```
